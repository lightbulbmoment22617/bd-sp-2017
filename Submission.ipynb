```python
prox_fixed = pd.read_csv("data/prox-fixed.csv")
prox_mobile = pd.read_csv("data/prox-mobile.csv")
bldg_measurements = pd.read_csv("data/bldg-measurements.csv")
f2z2_haz = pd.read_csv("data/f2z2-haz.csv")
```

```python
prox_fixed["timestamp"] = pd.to_datetime(prox_fixed["timestamp"])
prox_fixed[["date"]] = prox_fixed["timestamp"].apply(lambda x: x.date())
prox_mobile["timestamp"] = pd.to_datetime(prox_mobile["timestamp"])
prox_mobile[["date"]] = prox_mobile["timestamp"].apply(lambda x: x.date())
columns = ["date", " prox-id"]
pd.concat([prox_fixed[columns], prox_mobile[columns]]).groupby(["date"]).nunique()
```
