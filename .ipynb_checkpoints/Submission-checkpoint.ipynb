### Shell Command for Map Reduce

### Python Scripts

```python
%%script false --no-raise-error
# MAPPER

#!/usr/bin/env python3
import sys

for line in sys.stdin:
    line = line.strip()
    words = line.split()

    # skip first line
    if ("prox-id," in words):
        continue
    else:
        # (timestamp, id)
        print("%s\t%s" % (words[0], words[3]))
```

```python
%%script false --no-raise-error
# COMBINER

#!/usr/bin/env python3
import sys

current_day = None
current_prox_ids = []

for line in sys.stdin:
    line = line.strip()

    # output from mapper is (timestamp, prox_id)
    day, prox_id = line.split("\t", 1)

    if (day == current_day):
        # add the id to the list of ids for this day
        current_prox_ids.append(prox_id)
    else:
        if (current_day is not None):
            output = "%s\t" % current_day
            output += " ".join(map(str, current_prox_ids))
            # (timestamp [id, id, id, id, ...])
            print(output)

        current_day = day
        current_prox_ids = [prox_id]

# output final day
if (current_day is not None):
    output = "%s\t" % current_day
    output += " ".join(map(str, current_prox_ids))
    print(output)
```

```python
%%script false --no-raise-error
# REDUCER

#!usr/bin/env python3
import sys

current_day = None
current_prox_ids = []

for line in sys.stdin:
    line = line.strip()

    # output from combiner is (day, [list of prox_ids])
    day, prox_ids = line.split("\t", 1)
    prox_ids = prox_ids.split()

    if (day == current_day):
        current_prox_ids += prox_ids
    else:
        if (current_day is not None):
             # use python sets to quickly get the number of unique items
             # (date, number of ids)
             print("%s\t%s" % (current_day, len(set(current_prox_ids))))

        current_day = day
        current_prox_ids = prox_ids

if (current_day is not None):
    print("%s\t%s" % (current_day, len(set(current_prox_ids))))
```

### Shell Command for Map Reduce

### Python Scripts

```python
%%script false --no-raise-error
# MAPPER

#!/usr/bin/env python3
import sys

for line in sys.stdin:
    line = line.strip()
    words = line.split()

    # skip first line
    if ("prox-id," in words):
        continue
    else:
        # (floor zone, id)
        key = "%s %s" % (words[4], words[5])
        print("%s\t%s" % (key, 1))
```

```python
%%script false --no-raise-error
# REDUCER (ALSO USED AS COMBINER)

#!/usr/bin/env python3
import sys

current_fz = None
current_count = 0
max_fz = None
max_count = 0

def update_max(current_count, current_fz):
    global max_fz, max_count
    if (current_count > max_count):
        max_count = current_count
        max_fz = current_fz

for line in sys.stdin:
    line = line.strip()

    # output from mapper is (floor zone, 1)
    fz, count = line.split("\t", 1)

    # check input validity
    try:
        count = int(count)
    except ValueError:
        continue

    if (fz == current_fz):
        # update the count of number of visits
        current_count += count
    else:
        if (current_fz is not None):
            # see if we have a new "most visited" location
            update_max(current_count, current_fz)

        current_fz = fz
        current_count = count

if (current_fz):
    # check if we need to update the output first
    update_max(current_count, current_fz)
# print the most visited location
print("%s\t%s" % (max_fz, max_count))
```

### Shell Command for Map Reduce

```python
%%script false --no-raise-error
# MAPPER

#!/usr/bin/env python3
import sys

for line in sys.stdin:
    line = line.strip()
    words = line.split()

    # skip first line
    if ("prox-id," in words):
        continue
    # check if we have the correct date
    elif(words[0] == "2016-06-02"):
        # (id, 1)
        print("%s\t%s" % (words[3], 1))
```

```python
%%script false --no-raise-error
# REDUCER (ALSO USED AS COMBINER)

#!/usr/bin/env python3
import sys

current_id = None
current_count = 0
max_id = None
max_count = 0

def update_max(current_count, current_id):
    global max_id, max_count
    if (current_count > max_count):
        max_count = current_count
        max_id = current_id

for line in sys.stdin:
    line = line.strip()

    # output from mapper is (id , 1)
    new_id, count = line.split("\t", 1)

    # check input validity
    try:
        count = int(count)
    except ValueError:
        continue

    if (new_id == current_id):
        # update the count of number of visits
        current_count += count
    else:
        if (current_id):
            # might need to update maximum
            update_max(current_count, current_id)

        current_id = new_id
        current_count = count

# check maximum again and output most active id
if (current_id):
    update_max(current_count, current_id)

print("%s\t%s" % (max_id, max_count))
```

## Q4

### Answer

```python
# read output into arrays

all_hours = []
all_powers = []

file = open("coursework/q4/output.txt", "r")
for line in file:
    hour, power = line.split("\t", 1)
    all_hours.append(int(hour))
    all_powers.append(float(power))
```

```python
# plot the results

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(all_hours, all_powers, color="red", alpha=0.7, marker=".", markerfacecolor="black")
ax.set_xlabel("Hour")
ax.set_ylabel("Total Electric Demand Power (W)")
plt.title("Power Usage by Hour")
ax.set_xticks(all_hours)
plt.show()
```

The plot indicates that most of the power usage in the building occurs between the hours of 8am and 5pm - these are presumably the working hours. We see an increase in power usage from 6am and a decrease after 5pm which probably reflect the earliest arrivals to the building and latest departures respectively. For the remaining hours of the day, between 6pm and 6am we see the lowest level of power usage which makes sense as we would not expect anyone (other than perhaps security) to be in the building and using much power.

### Shell Command for Map Reduce

### Python Scripts

```python
%%script false --no-raise-error
# MAPPER

#!/usr/bin/env python3
import sys

for line in sys.stdin:
    line = line.strip()
    words = line.split()

    # skip first line
    if ("Date/Time," in words):
        continue
    else:
        # (hour, power)
        print("%s\t%s" % (words[1][:2], words[9][:-1]))
```

```python
%%script false --no-raise-error
# COMBINER

#!/usr/bin/env python3
import sys

current_hour = None
current_powers = []

for line in sys.stdin:
    line = line.strip()

    # output from mapper is (hour, power)
    hour, power = line.split("\t", 1)

    if (hour == current_hour):
        # add the power to the list of powers for this hour
        current_powers.append(power)
    else:
        if (current_hour is not None):
            output = "%s\t" % current_hour
            output += " ".join(map(str, current_powers))
            # (hour, [power, power, power, ...])
            print(output)

        current_hour = hour
        current_powers = [power]

# output final hour
if (current_hour is not None):
    output = "%s\t" % current_hour
    output += " ".join(map(str, current_powers))
    print(output)
```

```python
%%script false --no-raise-error
# REDUCER

#!usr/bin/env python3
import sys

current_hour = None
current_total = 0
current_count = 0

for line in sys.stdin:
    line = line.strip()

    # output from combiner is (hour, [list of power measurements])
    hour, powers = line.split("\t", 1)
    powers = powers.split()

    if (hour == current_hour):
        current_total += sum([float(x) for x in powers])
        current_count += len(powers)
    else:
        if (current_hour is not None and current_count > 0):
             average = current_total / current_count
             # (hour, average power)
             print("%s\t%s" % (current_hour, average))

        current_hour = hour
        current_total = sum([float(x) for x in powers])
        current_count = len(powers)

if (current_hour is not None and current_count > 0):
    average = current_total / current_count
    print("%s\t%s" % (current_hour, average))
```

## Q8

Writing Map Reduce programmes is certainly more cumbersome than their Spark equivalents. While I personally found it easier to implement logic in the first case I attribute this to my familiarity with Python (and the lack of it with Scala) as opposed to actual ease of use. The command to run Map Reduce on Hadoop is also huge while the same job can be run using Scala in a few lines.

The functional aspect of Scala also makes it easier to chain multiple maps/filters/etc than Map Reduce.

Lastly, I have noticed that the scalability of Hadoop is not very useful when working with datasets as small as these since running a Map Reduce program requires Hadoop to spin up several machines regardless of the dataset size. This in itself takes much much longer than my Scala programs did, since they were only really running anything when I wanted a specific output (lazy execution).

## Q10

We first plot the data to see if we can eyeball any association.

```python
# reading and prepping data

bldg_measurements = pd.read_csv("data/bldg-measurements.csv")
f2z2_haz = pd.read_csv("data/f2z2-haz.csv")
f2z2_haz.rename(columns={"Date/Time ": "Date/Time"}, inplace=True)
bldg_measurements["Date/Time"] = pd.to_datetime(bldg_measurements["Date/Time"])
f2z2_haz["Date/Time"] = pd.to_datetime(f2z2_haz["Date/Time"])
merged = f2z2_haz.merge(bldg_measurements, how="inner", on="Date/Time")[["Date/Time", " F_2_Z_1 VAV REHEAT Damper Position", " F_2_Z_2: Hazium Concentration"]]
merged.set_index("Date/Time", inplace=True)
```

```python
# plotting data

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1, 1, 1)

ax.scatter(merged.index, merged.iloc[:, 0], marker=".", color="r", linewidth=0.5, alpha=0.5, label="Damper Position")
ax.legend(loc=2)
ax.set_ylabel("Damper Position")

ax2 = ax.twinx()
ax2.plot(merged.iloc[:, 1], color="b", alpha=0.9, linewidth=0.5, label="Hazium")
ax2.legend(loc=1)
ax2.set_ylabel("Hazium Concentration")
ax.set_xlabel("Date")

plt.title("F2Z2 Damper Position and F2Z2 Hazium Concentration")
plt.show()
```

The plot above suggests that the Hazium concentration is positive at large values of the damper position. It seems that a damper position greater than (roughly) 0.8 corresponds to a positive Hazium concentration. To verify this we train a logistic regression classifier which uses the damper position to predict whether the Hazium concentration is positive.
